# Algorithmic trading and AI

In order to first clearly understand how AI has been applied to algorithmic trading, we need to briefly understand what AI itself is. AI or Artificial Intelligence is the simulation of human intelligence by machines in order to perform an array of tasks.

Algorithmic trading also known as automated trading is the process of the execution of trading operations which are deduced by an algorithm (Bajwa, 2022). As with time the algorithms have been improved, it has also resulted in traders gaining benefit from higher ROI’s (Return on Investment). In algorithmic trading there is little to less need for human intervention although it is needed to define the rules of the algorithm (Mallon, n.d.). Numerous top financial institutions are now switching their approach from normal trading to algorithmic trading due to its benefits.


## Why use AI in the algorithmic trading domain?

The strategy employed by algorithmic trading has provided traders with a lot of advantages. Here are a few,

### Elimination of human error
The ability of algorithmic trading to trade on a set of predetermined rules means that it eliminates the human psychological factors. Humans can make decisions based on false heuristics. AI can help in adapting to the new market trends, which is a time consuming process for humans and can lead to errors due to this (8topuz.com, n.d.).

### Speed of trades
The fact that AI based trading can automatically execute trades already gives a massive advantage to it over normal human involved trading. Algorithms can perform thousands of trades in fractions of seconds. This speed allows for more trading opportunities at even better prices (nasdaq.com, 2019).

### Accuracy of trades
Algorithmic trading performs trades automatically and this helps prevent from the human possibility of putting in a wrong trade and thereby creating errors. Instead, algorithmic trading performs all actions correctly at any point in time which makes them have greater accuracy.

### Ability to backtest
This is another great advantage of algorithmic trading. Backtesting allows companies to test their trading system based on past data and checking if the system would have been accurate on past data. This helps them evaluate the system before they make it live in order to start trading (nasdaq.com, 2019).

### Reduced cost
High speed of trades made by use of algorithmic trading allows for the same amount of profit made in a shorter period of time. Also the fact that it reduces the need for human supervision allows for saved costs (Mallon, n.d.).

### High Frequency Trading
Another form of algorithmic trading is known as High Frequency Trading (HFT). This form of trading is anticipated to be the most popular form of trading in the near future. The main concept is to buy and sell vast volumes of stocks mechanically. The same task if done by humans would take a lot of time due to the need to market research and manual bidding (Mishra, n.d.).


## Strategies used in algorithmic trading

There are several algorithmic trading strategies that are utilised today (Rafia, 2020),

### Trade Execution Algorithms
These algorithms split the trades into a bunch of smaller orders so that the impact on the stock price can be diminished.

### Strategy Implementation Algorithms
Trades are executed based on signals received from real time market analysis.

### Stealth/Gaming Algorithms
Other algorithmic strategies and huge trades can cause instability or movement in the market. These movements can be exploited by stealth or gaming algorithms.

### Arbitrage Opportunities
This strategy is basically possible when the same stock is listed on two different markets. In such a scenario, stocks are sold on the market that has the higher price and bought on the market that has the lower price.


## Problems associated with using AI in trading

Now that we have looked at the positives of using AI in the algorithmic trading industry, it is also worth mentioning the negative side (Singh, 2021). 

### Interpretability 

Some AI algorithms used for trading are not predictable and cannot be easily explained. An algorithm might produce an outcome but in certain algorithms it is difficult to explain why the algorithm came to a certain conclusion (a-teaminsight.com, 2018). 

### Market volatility factor

Models trained in periods of low market volatility might not be an appropriate choice to be used in periods of high market volatility.

### Data quality

An important factor which makes a good algorithmic trading system is the data quality that is used to train the system. Even if there are a few observations of corrupted data, it can heavily influence the performance of  the system and potentially lead to critical consequences.

### Data bias

Another challenge is the existence of biassed data, a decision made by an AI can have a massive impact on the subject. This is why there is a need to remove any source of bias in the data.

### Long Short Term Memory (LSTM)

LSTM is generally a variety of recurrent neural networks that are capable of learning long term dependencies in sequence prediction problems. LSTM makes use of feedback connections. Some applications of LSTM can be speech recognition and machine translation (intellipaat.com, 2022).

LSTM consists of three parts which each on their own perform a specific function. The first part makes a decision upon whether the information coming from the previous cell is irrelevant and can be forgotten. The second part consists of the cell learning new information from the current input. In the third part the cell passes the new information to the next timestamp from the current timestamp.



In the algorithmic trading domain, LSTM can be used to create reliable stock price predictions systems that can help traders to make better decisions. For example, traders can look at the price of a certain stock within the next 30 days and accordingly decide whether to invest or not.

<u>Strengths</u>

According to Budiharto (Budiharto, 2021), here are some of the strengths of LSTM

i) LSTM is able to capture long-term dependencies in data, which can be important for accurately predicting stock prices.

ii) LSTM is able to handle variable-length sequences, which is useful in the case of stock price prediction where the length of the time series may vary depending on the stock and the period of time being considered.

iii) LSTM is able to solve the vanishing and exploding gradients problem by using a gating mechanism that controls the flow of information through the network. This gating mechanism consists of input, forget, and output gates that are used to regulate the flow of information in and out of the LSTM cell.

<u>Weaknesses</u>

According to geeksforgeeks.com, here are a few drawbacks of using LSTM

i) It turns out that LSTM cannot completely solve the problem of vanishing gradients.

ii) According to Leccese (Leccese, 2019), LSTM only provides a predicted value that is very close to the previous value. Hence, they believe that LSTM has no predictive capability.

iii) LSTM is computationally intensive, which can make it difficult to train and deploy on large datasets or in real-time applications.

iv) LSTM may not always be the best choice for stock price prediction, depending on the specific characteristics of the data and the desired prediction accuracy. Other machine learning models, such as support vector machines or random forests, may perform better in some cases.

### Genetic Algorithms

Genetic algorithms are a type of optimization algorithm that are used in machine learning. They work by simulating the logic of Darwanian selection where only the best are chosen. They are used to solve complex problems which have numerous variables and solutions. Different solutions are run through these algorithms to find the best set of solutions (vadapalli, 2020). The phases of the working of such optimization tasks using genetic algorithms includes “Initialization, Fitness assignment, Selection, Reproduction, Replacement, and Termination” (Muthee, 2021). 



According to mindmajix.com; There are several advantages and disadvantages to using genetic algorithms, here are some of them listed below,

#### <u>Advantages</u>

i) Derivative information is not necessary for genetic algorithms

ii) When compared to the conventional methods of brute-force search, genetic algorithms are more rapid and effective.

iii) There is evidence that genetic algorithms offer numerous parallel capabilities.

iv) Optimises multi-objective problems as well as continuous and discrete functions.

#### <u>Disadvantages</u>

i) For simpler issues when the derivative information is easily available, genetic algorithms are not the most effective solution.

ii) When utilising Genetic Algorithms, the fitness value is evaluated across a number of generations, which can be a costly process for some problems.

iii) A genetic algorithm may not converge to the best answer if it is not used properly.

# Implementation 

In this section we will describe the input required for this implementation. Where will we get this data from and comment on the formatting needs.

### Description of the input
The purpose of this implementation is to create an AI system for making a stock price prediction model using LSTM recrruent neural networks. Since stock price prediction models require historical data, the data we will provide our LSTM model will the time series data. Time series data is a type of sequential data that is collected over time and ordered by time.

### Source of data
The data to be used for training our model will come from the python package 'yfinance'. Yfinance is a python package that help us retrieve stock market data from yahoo finance. It can provide us with historical stock price of many publicly traded companies.

### Formatting needs
We need to do the basic preprocessing like removing missing values firstly. The data also needs to be scaled first as LSTM is sensitive to the scale of the input data. After perfroming the train test split, we then need to create a set of input sequences, where each sequence consists of a series of time steps. LSTM needs to be fed data in a 3D array, where the three dimension are sampes, timesteps, and features thus the data needs to be finally reshaped and then fed to the LSTM model.

Here is a high level diagram portraying how different elements in the project work together



Below is code implementation of the project

```python
import yfinance as yf
from datetime import datetime
import numpy as np
```

```python
# retrieve stock information 
tickers = yf.Ticker("AMZN")

# set end date to current date
ending_date = datetime.now().strftime('%Y-%m-%d')

# stock data from start of 2021 til now with a one day interval
stock_df = tickers.history(start='1970-01-01',end=ending_date,interval='1D')

# view the obtained dataframe
stock_df
```

```python
stock_df.shape
```

```python
# If there are any rows with missing data, then remove these rows
stock_df = stock_df.dropna()

# Fill in missing values with the mean
stock_df = stock_df.fillna(stock_df.mean())
```

```python
# view first five records of dataframe after dealing with missing data
stock_df.head()
```

```python
# retrieve the 'Close' column and reset index of the outputted dataframe
close_df = stock_df.reset_index()['Close']
```

```python
# view head of obtained dataframe of column 'Close'
close_df.head()
```

```python
# import matplotlib.pyplot and plot the close variable to see how it changes
import matplotlib.pyplot as plt
plt.plot(close_df)
```

```python
# import relavant packages for scaling data and building lstm model
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
```

```python
# Preprocess the data, as lstm is sensitve and needs the data to be on the same scale

# scale the features within a range of 0 and 1
scaler = MinMaxScaler(feature_range=(0,1))
# The data is scaled by fitting the scaler on the reshaped 2D numpy array
stock_data = scaler.fit_transform(np.array(close_df).reshape(-1,1))
```

```python
# Training and test split, the training data is given a size of 75% while the rest can be used by test
train_size = int(len(stock_data) * 0.75)
test_size = len(stock_data) - train_size

# scaled data array sliced to make train and test subsets.
train_data, test_data = stock_data[0:train_size, :], stock_data[train_size:len(stock_data), :1]
```

```python
# This function creates two numpy arrays, the first one contains the data for the defined number of timesteps and the second
# - specifies the target value
def X_Y_train_test_split(data, previous_steps=1):
    X_data, y_data = [], []
    for i in range(len(data)-previous_steps-1):
        row_values = data[i:(i+previous_steps), 0]
        X_data.append(row_values)
        y_data.append(data[i+previous_steps, 0])
    return np.array(X_data), np.array(y_data)
```

```python
previous_steps = 20
X_train, y_train = X_Y_train_test_split(train_data, previous_steps)
X_test, y_test = X_Y_train_test_split(test_data, previous_steps)
```

```python
print(X_train.shape), print(y_train.shape)
```

```python
print(X_test.shape), print(y_test.shape)
```

```python
#reshaping the data to a 3D format as LSTM requires data in 3-Dimensional space
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)
```

```python
# Build the LSTM model
model = Sequential()
model.add(LSTM(200, return_sequences=True,input_shape=(X_train.shape[1],X_train.shape[2])))
model.add(LSTM(200, return_sequences=True))
model.add(LSTM(50))
model.add(Dense(25))
model.add(Dense(1))
```

```python
# Compile the model
model.compile(loss="mean_squared_error", optimizer="adam")
```

```python
model.summary()
```

```python
# Train the model with specified parameters
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=100, batch_size=64, verbose=1)
```

```python
# predict the values for X_test and X_train
train_predict=model.predict(X_train)
test_predict=model.predict(X_test)
# reverse the scaling to use rmse as evaluation metric
train_predict=scaler.inverse_transform(train_predict)
test_predict=scaler.inverse_transform(test_predict)
```

```python
import math
from sklearn.metrics import mean_squared_error
```

```python
# new duplicated reshaped y_test, y_train to calculate rmse
y_train_1 = scaler.inverse_transform(y_train.reshape(-1,1)) 
y_test_1 = scaler.inverse_transform(y_test.reshape(-1,1))
```

```python
# RMSE between predicted training set values and actual training target values
math.sqrt(mean_squared_error(y_train_1, train_predict))
```

```python
# RMSE between predicted test set values and actual test target values
math.sqrt(mean_squared_error(y_test_1, test_predict))
```

```python
# plot of predicted test values
plt.plot(test_predict)
```

```python
# plot of actual test values
plt.plot(y_test)
```

There a various testing methods that can be used for a stock price prediction model using LSTM. Here a couple of them defined below,

##### <u>Holdout validation</u>
This includes creating a training and test set for your data. A model is trained on the training set and evaluated on test set to check how well the model generalize to unseen data points (Allibhai, 2018).
##### <u>K-fold cross validation</u>
Here, the data is split into k-groups. The model is tested on one of the sets and trained on the rest. This process repeats k number of times so that each fold has been used as a test set once (Allibhai, 2018).

For this prototype we will be using Holdout validation for simplicity.
